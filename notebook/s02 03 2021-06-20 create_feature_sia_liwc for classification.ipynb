### get clean raw data

```python
clean_data_dir = os.path.abspath(os.path.join(os.pardir, 'data', 's2021_lumen_clean_data'))
cd_sia_liwc_file = 's2021_05_23_03_lumen_clean_doc_sia_liwc.csv'
cd_sia_liwc_path = os.path.join(clean_data_dir, cd_sia_liwc_file)
```

```python
cd_df = pd.read_csv(cd_sia_liwc_path)

cd_df
```

### clean up for classification

- remove doc has no labeling
- remove pre-processed nostop-stemmed emails with less than 10 words

```python
cd_df.columns
```

```python
label_clms = [
    'Authority or Expertise/Source Credibility',
    'Blame/guilt',
    'Commitment',
    'Commitment- Call to Action',
    'Commitment- Indignation',
    'Emphasis',
    'Gain framing',
    'Liking',
    'Loss framing',
    'Objectivity',
    'Reciprocation',
    'Scarcity/Urgency/Opportunity',
    'Social Proof',
    'Social Proof- Admonition',
    'Subjectivity',
]

len(label_clms)
```

```python
idx1 = cd_df.nostop_stem_doc_len >= 10
sum(-idx1)
```

```python
idx2 = cd_df[label_clms].any(axis=1)
sum(-idx2)
```

```python
cd_df.shape
```

```python
cd_class_df = cd_df[idx1 & idx2].copy()
cd_class_df.shape
```

### save data

```python
cd_sia_liwc_class_file = 's2021_06_20_01_lumen_clean_doc_sia_liwc_classify.csv'
cd_sia_liwc_class_path = os.path.join(clean_data_dir, cd_sia_liwc_class_file)
cd_class_df.to_csv(cd_sia_liwc_class_path, index=False)
```
